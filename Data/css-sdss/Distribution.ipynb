## Distribution

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

%matplotlib inline
```

```python
x_match_path = './X-match/keys.csv'
```

```python
def distribution(fig, axes, path, colors, norm=True, repeated = True):
    axes = fig.add_subplot(axes)
    x_match = pd.read_csv(path)
    
    if not repeated:
        x_match.drop_duplicates(subset=['css_num_ID'],inplace=True)
    
    count = x_match['label'].value_counts()
    
    y = count.values
    y_norm = y/sum(y)
    x = list(range(len(count.values)))
    xticks = count.index
    colors = colors[:len(x)]
    
    if not norm:        
        plt.bar(x=x, height=y, color=colors)
    else:
        plt.bar(x=x, height=y_norm, color=colors)
    
    title = 'Dataset distribution\n Normed: '+str(norm)+' Repeated: '+str(repeated)
    axes.set_xticklabels(xticks, rotation=60)
    plt.ylabel('Count')
    plt.title(title)
```

```python
plt.rcdefaults()
colors = sns.color_palette('tab10')
fig = plt.figure(figsize=(10,8), dpi=150)
distribution(fig, 221, x_match_path, colors, repeated=True, norm=False)
distribution(fig, 222, x_match_path, colors, repeated=True, norm=True)
distribution(fig, 223, x_match_path, colors, repeated=False, norm=False)
distribution(fig, 224, x_match_path, colors, repeated=False, norm=True)
plt.subplots_adjust(hspace=0.4)
```
